```python
def haversine(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance in kilometers between two points 
    on the earth (specified in decimal degrees)
    """
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])

    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 6371 # Radius of earth in kilometers. Use 3956 for miles. Determines return value units.
    return c * r

```

```python
df_cc_data['distance_km'] = df_cc_data.apply(lambda row: haversine(row['long'], row['lat'], row['merch_long'], row['merch_lat']), axis=1)
df_cc_data['distance_km'] = df_cc_data['distance_km'].round(2)
```

```python
fraud_rate_by_hour = df_cc_data.groupby('hour')['is_fraud'].mean()

plt.figure(figsize=(10, 6))
fraud_rate_by_hour.plot(kind='bar', color='orange')
plt.title('Fraud Rate by Hour of Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Fraud Rate')
plt.show()
```

```python
fraud_rate_by_hour = df_cc_data.groupby('hour')['is_fraud'].mean()
fraud_std_by_hour = df_cc_data.groupby('hour')['is_fraud'].std()
fraud_count_by_hour = df_cc_data.groupby('hour')['is_fraud'].count()

# Standard error
fraud_se_by_hour = fraud_std_by_hour / np.sqrt(fraud_count_by_hour)

plt.figure(figsize=(10, 6))
plt.bar(fraud_rate_by_hour.index, fraud_rate_by_hour, yerr=fraud_se_by_hour, color='orange', capsize=5)
plt.title('Fraud Rate by Hour of the Day with Confidence Intervals')
plt.xlabel('Hour of the Day')
plt.ylabel('Fraud Rate')
plt.show()
```

***Statistical Analysis***

```python

fraud = df_cc_data[df_cc_data['is_fraud'] == 1]
non_fraud = df_cc_data[df_cc_data['is_fraud'] == 0]

t_stat, p_value = ttest_ind(fraud['amt'], non_fraud['amt'])
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

```python
from scipy.stats import mannwhitneyu

fraud_amt = df_cc_data[df_cc_data['is_fraud'] == 1]['amt']
non_fraud_amt = df_cc_data[df_cc_data['is_fraud'] == 0]['amt']

stat, p_value = mannwhitneyu(fraud_amt, non_fraud_amt)
print(f"Mann-Whitney U Test: Statistic={stat}, p-value={p_value}")
```
